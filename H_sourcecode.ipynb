## Data loading

### Subtask:
Load the data from "RGB_colors.csv" into a pandas DataFrame.


**Reasoning**:
Load the data from "RGB_colors.csv" into a pandas DataFrame and display the first 5 rows.



```python
import pandas as pd

try:
    df = pd.read_csv('RGB_colors.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: 'RGB_colors.csv' not found.")
    df = None
except pd.errors.ParserError:
    print("Error: Could not parse the CSV file. Please check its format.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

## Data exploration

### Subtask:
Explore the loaded dataset to understand its characteristics.


**Reasoning**:
Examine the shape, data types, missing values, distribution of RGB values, and correlations within the DataFrame.



```python
# Examine the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Check data types
print("\nData Types:\n", df.dtypes)

# Identify missing values
print("\nMissing Values per column:\n", df.isnull().sum())

# Analyze the distribution of RGB values
print("\nDescriptive Statistics for RGB values:\n", df[['196', '98', '16']].describe())

# Initial Correlation Check
print("\nCorrelation between RGB values:\n", df[['196', '98', '16']].corr())
```

## Data visualization

### Subtask:
Visualize the distribution of RGB values and their relationships.


**Reasoning**:
Create histograms for each RGB component and a scatter plot matrix to visualize their distributions and relationships.



```python
import matplotlib.pyplot as plt

# Create histograms for each RGB component
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(df['196'], bins=20, color='red', alpha=0.7)
plt.xlabel('R Value')
plt.ylabel('Frequency')
plt.title('Distribution of R values')

plt.subplot(1, 3, 2)
plt.hist(df['98'], bins=20, color='green', alpha=0.7)
plt.xlabel('G Value')
plt.ylabel('Frequency')
plt.title('Distribution of G values')

plt.subplot(1, 3, 3)
plt.hist(df['16'], bins=20, color='blue', alpha=0.7)
plt.xlabel('B Value')
plt.ylabel('Frequency')
plt.title('Distribution of B values')

plt.tight_layout()
plt.show()


# Create scatter plot matrix
from pandas.plotting import scatter_matrix
plt.figure(figsize=(10, 10))
scatter_matrix(df[['196', '98', '16']], alpha=0.8, figsize=(10, 10), diagonal='hist')
plt.show()
```
